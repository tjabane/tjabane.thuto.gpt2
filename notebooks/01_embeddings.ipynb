# Embeddings
Embedding is transforming t text into a numerical representation that the model can work with. Embeddings are generated as following:

1. Tokenize the input string
2. Convert tokens into embeddings (vectors)
3. Add positional information

```python
import tiktoken
import torch
```

## 1. Tokenization

```python
tokenizer = tiktoken.get_encoding("gpt2")
sample_token = tokenizer.encode("Journey before destination.")
print(f"Sample token: {sample_token}")
print(f"Decoded sample token: {tokenizer.decode(sample_token)}")
```

## 2. Create Embeddings

```python
vocab_size = tokenizer.n_vocab
embedding_length = 128
embedding_layer = torch.nn.Embedding(num_embeddings=vocab_size, embedding_dim=embedding_length)
print(f"Embedding layer: {embedding_layer}")
```

## 3. Positional Encoding

### Absolute Positional Encoding
Absolute positional embedding is like giving each token in a sentence a unique number.
 $$PE_{(pos, 2i)} = \sin\left(\frac{pos}{10000^{2i/d_{\text{model}}}}\right), \qquad
  PE_{(pos, 2i+1)} = \cos \left(\frac{pos}{10000^{2i/d_{\text{model}}}}\right)$$
